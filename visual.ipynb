```python
df_l = pd.read_csv('learn.txt', sep=';')
```

```python
df_l
```

```python
width = 20
height = 10
plt.figure(figsize=(width, height))

fg1 = plt.subplot(3,4,1)
fg1.set_title('Standard')
plt.plot(df_l['Standard'], color='red')

fg5 = plt.subplot(3,4,5)
fg5.set_title('Standard_GSS')
plt.plot( range(0, df_l.count()['Standard']), df_l['Standard_GSS'], color='red')

maximum = df_l['Standard'].max() if df_l['Standard'].max() > df_l['Standard_GSS'].max() else df_l['Standard_GSS'].max()

fg9 = plt.subplot(3,4,9)
fg9.set_title('Standard(Blue) / Standard_GSS(Green)')
fg9.set_ylim([0, 0.1])
plt.plot(range(0,df_l.count()['Standard']), df_l['Standard']/maximum,
         range(0,df_l.count()['Standard']), df_l['Standard_GSS']/maximum)

fg2 = plt.subplot(3,4,2)
fg2.set_title('Threshold')
plt.plot(df_l['Threshold'])

fg6 = plt.subplot(3,4,6)
fg6.set_title('Threshold_GSS')
plt.plot(df_l['Threshold_GSS']) 

maximum = df_l['Threshold'].max() if df_l['Threshold'].max() > df_l['Threshold_GSS'].max() else df_l['Threshold_GSS'].max()

fg10 = plt.subplot(3,4,10)
fg10.set_title('Threshold(Blue) / Threshold_GSS(Green)')
fg10.set_ylim([0, 0.4])
plt.plot(range(0,df_l.count()['Threshold']), df_l['Threshold']/maximum,
         range(0,df_l.count()['Threshold']), df_l['Threshold_GSS']/maximum)

fg3 = plt.subplot(3,4,3)
plt.plot(df_l['FaultInjection'], color='green')
fg3.set_title('Fault Injection')

fg7 = plt.subplot(3,4,7)
plt.plot(df_l['FaultInjection_GSS'], color='green')
fg7.set_title('Fault Injection GSS')

maximum = df_l['FaultInjection'].max() if df_l['FaultInjection'].max() > df_l['FaultInjection_GSS'].max() else df_l['FaultInjection_GSS'].max()

fg11 = plt.subplot(3,4,11)
fg11.set_title('FaultInjection(Blue) / FaultInjection_GSS(Green)')
fg11.set_ylim([0, 0.05])
plt.plot(range(0,df_l.count()['FaultInjection']), df_l['FaultInjection']/maximum,
         range(0,df_l.count()['FaultInjection']), df_l['FaultInjection_GSS']/maximum)

fg4 = plt.subplot(3,4,4)
plt.plot(df_l['NoiseInjetion'], color='purple')
fg4.set_title('Noise Injection')

fg8 = plt.subplot(3,4,8)
plt.plot(df_l['NoiseInjetion_GSS'], color='purple')
fg8.set_title('Noise Injection GSS')

maximum = df_l['NoiseInjetion'].max() if df_l['NoiseInjetion'].max() > df_l['NoiseInjetion_GSS'].max() else df_l['NoiseInjetion_GSS'].max()

fg12 = plt.subplot(3,4,12)
fg12.set_title('NoiseInjetion(Blue) / NoiseInjetion_GSS(Green)')
fg12.set_ylim([0, 0.05])
plt.plot(range(0,df_l.count()['NoiseInjetion']), df_l['NoiseInjetion']/maximum,
         range(0,df_l.count()['NoiseInjetion']), df_l['NoiseInjetion_GSS']/maximum)
```

```python
width = 20
height = 10
plt.figure(figsize=(width, height))

fg1 = plt.subplot(1,3,1)
fg1.set_title('NeuralField')
fg1.plot(df_l['NeuralField'], color='red')

fg2 = plt.subplot(1,3,2)
fg2.set_title('NeuralField_GSS')
fg2.plot(df_l['NeuralField_GSS'])

maximum = df_l['NeuralField'].max() if df_l['NeuralField'].max() > df_l['NeuralField_GSS'].max() else df_l['NeuralField_GSS'].max()

fg3 = plt.subplot(1,3,3)
fg3.set_title('NeuralField(Blue) / NeuralField_GSS(Green)')
plt.plot(range(0,df_l.count()['NeuralField']), df_l['NeuralField']/maximum,
         range(0,df_l.count()['NeuralField']), df_l['NeuralField_GSS']/maximum)
```
